```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("breast-cancer.csv")
df
```

```python
data = df.iloc[:,1:]
data
```

```python
# Benign: The tumor doesn’t contain cancerous cells.
# Malignant: The tumor contains cancerous cells.
data["diagnosis"] = np.where(data["diagnosis"] == "B", "Benign", "Malignant")
data
```

```python
data.info() # show data type and more information
```

```python
data.isna().sum() # check is any NA value exist in data or not
```

```python
data.drop_duplicates(inplace = True) # drop duplicates value if it exist
data
```

```python
# Use corr() function to find the correlation among the columns in the Dataframe

plt.figure(figsize = (25,16))
sns.heatmap(data.corr(), annot = True, linewidths = 1, cmap = "YlGnBu")
plt.tight_layout()
plt.show()

# Score > 0.85, strong correlation (Dark Blue)
# Score 0.4<>0.85, moderate correlation (between Dark blue and light blue)
# Score <0.4, weak correlation
```

```python
#### check output column is Normal or Not with atleast 70:30 ratio
label = data["diagnosis"].value_counts().keys()
size = data["diagnosis"].value_counts().values

plt.pie(size, labels = label, autopct = "%1.0f%%", explode=[0, 0.05])
plt.xlabel("diagnosis")
plt.legend()
plt.show()
```
